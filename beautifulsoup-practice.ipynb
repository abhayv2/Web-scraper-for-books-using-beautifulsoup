```
URL = "https://seenunseen.in/episodes/2021/"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')

data = {}
for row in  soup.findAll('h2', attrs = {'class':'entry-title'}):
  episode = row.text
  link = row.a['href']
  NewUrl = link

  # For each episode
  book=[]
  r = requests.get(NewUrl)
  soup = BeautifulSoup(r.content, 'html5lib')
  for row in soup.findAll('a'):
    if "www.amazon.in" in str(row):
      book.append(row.text)
  data.update({episode:book})


```
